```python
import xgboost
import shap
import seaborn as sns
```

```python
# train an XGBoost model with california housing data
X, y = shap.datasets.california()
model = xgboost.XGBRegressor().fit(X, y)
```

```python
X.describe()
```

```python
X.columns
```

```python
# This will create a visual showing the relationship among pollutants.
grid_max = sns.pairplot(X,
                        plot_kws=dict(linewidth=0.7),
                        diag_kind='kde',height=2,corner=True,markers="+",
                        diag_kws=dict(linewidth=0.7));
grid_max;
```

```python
# explain the model's predictions using SHAP
# (same syntax works for LightGBM, CatBoost, scikit-learn, transformers, Spark, etc.)
explainer = shap.Explainer(model)
shap_values = explainer(X)

# Choose an observation for which you want to visualize the waterfall plot
observation_index = 0

# visualize the first prediction's explanation
shap.plots.waterfall(shap_values[observation_index])
```

```python
shap_values[0].shape
```

 Higher predictions are given by the ones shown in red and the ones in blue push to lower predictions. 

```python
# visualize the first prediction's explanation with a force plot
shap.initjs()
shap.plots.force(shap_values[0])
```

```python
# visualize all the training set predictions
shap.plots.force(shap_values[:500])
```

```python
# Variation in predicted house price as the latitude is modified
# create a dependence scatter plot to show the effect of a single feature across the whole dataset
shap.plots.scatter(shap_values[:, "Latitude"], color=shap_values)
```

```python
shap.plots.scatter(shap_values[:, "MedInc"], color=shap_values) #Median Income
```

```python
# summarize the effects of all the features
shap.plots.beeswarm(shap_values)
```
